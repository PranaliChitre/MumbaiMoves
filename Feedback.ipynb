```python
import matplotlib.pyplot as plt
from textblob import TextBlob

def analyze_sentiment_textblob(text):
    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Suggestion'

# Example usage with Mumbai-centric texts
texts_to_analyze = ["The BEST bus frequency is quite good in my area",
                    "The cleanliness of Mumbai buses needs improvement",
                    "Need more AC buses for long routes in Mumbai",
                    "Mumbai bus drivers are very courteous and helpful",
                    "The ticketing system of Mumbai buses needs to be modernized",
                   "Not a good experience"]

sentiment_results = [analyze_sentiment_textblob(text) for text in texts_to_analyze]

# Plotting a bar chart
sentiment_counts = {'Positive': 0, 'Negative': 0, 'Suggestion': 0}

for result in sentiment_results:
    sentiment_counts[result] += 1

labels = list(sentiment_counts.keys())
values = list(sentiment_counts.values())

plt.bar(labels, values, color=['green', 'red', 'gray'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Analysis Results of Mumbai Bus Services')
plt.show()

# Displaying sentiment for each example
for i, text in enumerate(texts_to_analyze):
    sentiment_result = analyze_sentiment_textblob(text)
    print(f"Example {i+1}: '{text}' - Sentiment: {sentiment_result}")
```
